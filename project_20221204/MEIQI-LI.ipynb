## Table 1: International Migrant Stock at Mid-Year

```python
import numpy as np
import pandas as pd
```

```python
UN = '../project data/UN_MigrantStockTotal_2015.xlsx'
dataframes = pd.read_excel(UN, sheet_name=None, header=[15])

dataframes.keys()
```

```python
t1 = dataframes["Table 1"]
t1.head()
```

```python
# Rename columns with appropriate title
t1.columns = ['Order', 'Destination', 'Notes', 'Country code', 'Data type', 
                   '1990B','1995B','2000B','2005B','2010B','2015B',
                   '1990M','1995M','2000M','2005M','2010M','2015M',
                   '1990F','1995F','2000F','2005F','2010F','2015F'
                   ]

t1.replace('..',np.nan,inplace = True) # Replace ".." with NaN

t1.drop(columns=['Notes', 'Data type'], inplace = True) # Remove unrelevant data

t1
```

```python
# Problem: Column names are values not variable names
# tidy data principle #1 Column names need to be informative, variable names and not values

t1 = t1.melt(id_vars=["Order","Destination","Country code"], var_name = "Year", value_name = "International Migrant Stock")

t1
```

```python
# Problem: there are multiple variables stored in 1 column
# tidy data principle #2: each column needs to consist of one and only one variable

# Add Gender column 
t1 = t1.assign(Gender = lambda x: x['Year'].str[-1])
t1['Gender'] = t1['Gender'].apply(lambda x: "Male" if (x == "M") else ("Both Sexes" if (x == "B") else "Female"))

# Remove gender in Year column
t1['Year'] = t1['Year'].str[:-1].astype(str)
t1
```

```python
#Problem: Variables are stored in both rows and columns
# tidy data principle #3 Variables need to be in cells, not rows and columns
t1.pivot_table(index = ["Order", "Destination","Country code","Year"], columns = "Gender", values = "International Migrant Stock").reset_index().rename_axis(None, axis=1).replace(-1, np.nan)
```

## Table 6: Estimated Refugee Stock at Mid-Year

```python
t6 = dataframes['Table 6']
t6
```

```python
# Rename columns with appropriate title


t6.columns = ['Order', 'Destination', 'Notes', 'Country code', 'Type of data',
                     '1990E','1995E','2000E','2005E','2010E','2015E',
                     '1990P','1995P','2000P','2005P','2010P','2015P',
                     '1990-1995A','1995-2000A','2000-2005A','2005-2010A','2010-2015A'
                   ]

t6.replace('..',np.nan,inplace = True) # Replace ".." with NaN
t6
```

```python
t6.drop(['Notes', 'Type of data'],axis=1,inplace=True) # Remove unrelevant data

t6
```

```python
# Problem: There are multiple types of data stored in one table
# tidy data principle #4 Each table coloumn needs to have a singular data type
t6 = t6.set_index(["Order","Destination","Country code"])
refugees_stock = t6[['1990E', '1995E', '2000E', '2005E', '2010E', '2015E']].copy()
refugees_percentage = t6[['1990P', '1995P', '2000P', '2005P', '2010P', '2015P']].copy()
rate_of_change = t6[['1990-1995A', '1995-2000A', '2000-2005A', '2005-2010A', '2010-2015A']].copy()
```

```python
# Problem: Column names are values not variable names
# tidy data principle #1 Column names need to be informative, variable names and not values
refugees_stock = refugees_stock.reset_index().melt(id_vars =  [ 'Order', 'Destination','Country code',]
                       , var_name = "Year", value_name = "Estimated refugees stock at mid-year")
# Clean Year column
refugees_stock['Year'] = refugees_stock['Year'].str[:-1].astype(str)

refugees_stock
```

```python
refugees_percentage = refugees_percentage.reset_index().melt(id_vars =  [ 'Order', 'Destination','Country code',]
                       , var_name = "Year", value_name = "Refugees as a percentage of the international migrant stock")
# Clean Year column
refugees_percentage['Year'] = refugees_percentage['Year'].str[:-1].astype(str)

refugees_percentage
```

```python
rate_of_change = rate_of_change.reset_index().melt(id_vars =  [ 'Order', 'Destination','Country code',]
                       , var_name = "Year", value_name = "Annual rate of change of the refugee stock")
# Clean Year column
rate_of_change['Year'] = rate_of_change['Year'].str[:-1].astype(str)

rate_of_change
```

```python
with pd.ExcelWriter('./tidy_tables.xlsx') as writer:  
    t3.to_excel(writer, sheet_name='Migrant Stock and Population')
    t4.to_excel(writer, sheet_name='Female and Refugee Percentage')
    rate_of_change.to_excel(writer, sheet_name='Annual Rate of Change')
    refugees_stock.to_excel(writer, sheet_name='Estimate Refugee Stock')
writer.save()
```
