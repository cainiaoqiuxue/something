```python
import pandas as pd
```

```python
df=pd.read_excel('data.xlsx').iloc[:,1:]
```

```python
df.head()
```

```python
df=df[df['time'].str.contains('\d{2}-\d{2}',regex=True)]
df=df[~df['read'].str.contains('万')]
df['read']=df['read'].astype(int)
df['comment']=df['comment'].astype(int)
df['year']=''
df.iloc[:38625,5]='2020'
df.iloc[38625:,5]='2019'
df['date']=df.apply(lambda x:x.year+'-'+x.time[:5],axis=1)
df.iloc[38620:38630,:]
```

```python
res=df.groupby('date')[['read','comment']].agg(['count','sum']).sort_index()
```

```python
sub=pd.DataFrame({
    'date':res.index,
    'count':res['read']['count'],
    'read':res['read']['sum'],
    'comment':res['comment']['sum']
})
```

```python
sub.to_excel('final.xlsx',index=False)
```
