```python
import pandas as pd
from src.process.data_process import sentiment_analysis
```

```python
df = pd.read_excel('./data/新疆棉中文_4675.xlsx')
```

```python
df.fillna('', inplace=True)
```

```python
df['text'] = df.apply(lambda x: x['标题'] + x['摘要'] + x['正文'], axis=1)
```

```python
sentiment_analysis(df, './models/distilbert-base-multilingual-cased-sentiments-student')
```

```python
df['情感倾向'] = df['情感倾向'].map({0: '情感正向', 1: '情感中立', 2: '情感负向'})
```

```python
df['情感倾向'].value_counts().to_dict()
```

```python
df.to_excel('./data/zhanshi.xlsx', index=False)
```

```python
df[df['情感倾向'] == 2]['text'].sample(5).tolist()
```

```python
source_dict = {}
for k,v in df['信息来源'].value_counts().to_dict().items():
    if ('网' in str(k) or '新闻' in str(k)) and v > 10:
        source_dict[k] = v
```

```python
source_dict['其他'] = df.shape[0] - sum(source_dict.values())
```

```python
df['发布时间'] = df['发布时间'].apply(lambda x: str(x)[:10])
```

```python
data = df.groupby(['发布时间', '情感倾向']).count()['text'].reset_index()
```
