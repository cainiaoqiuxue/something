```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
train = pd.read_csv('New folder/C_4_train.csv')
test = pd.read_csv('New folder/C_4_test.csv')
```

```python
sns.histplot(train['revenue'])
sns.kdeplot(train['revenue'])
```

```python
sns.kdeplot(np.log(train['revenue']))
sns.histplot(np.log(train['revenue']), kde=True)
```

```python
test['revenue'] = -1
```

```python
df = pd.concat([train, test], axis=0, ignore_index=True)
```
