```python
import os
import pandas as pd

os.chdir('../data/选题二训练集：热点事件的发展趋势预测/')

def get_data_id(tid):
    try:
        df = pd.read_excel('数据/{}.xlsx'.format(tid))
    except:
        df = None
    return df


def get_stats_value(row):
    tid = row['序号']
    df = get_data_id(tid)
    if df is None:
        return None, None, None, None, None, None, None
    num = df.shape[0]
    dz = df['点赞'].mean()
    zf = df['转发'].mean()
    pl = df['评论'].mean()
    fs = df['粉丝数'].mean()
    gz = df['关注数'].mean()
    dy = df['地域'].nunique()
    return num, dz, zf, pl, fs, gz, dy

```

```python
df = pd.read_excel('数据集序号和事件对应表.xlsx')
df1 = pd.read_excel('数据/1.xlsx')
```

```python
tmp = df.apply(get_stats_value, axis=1, result_type='expand')

df['微博数 点赞 转发 评论 粉丝数 关注数 地域数'.split(' ')] = tmp
```
